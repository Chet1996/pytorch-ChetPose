```python
import sys
sys.path.append('..')
from transforms import *
import matplotlib.pyplot as plt
%matplotlib inline
from config import config
from utils import display_one_img
```

```python
json_file = os.path.join('..', config.train_json_file)
img_dir = os.path.join('..', config.train_img_dir)
json_file, img_dir
```

```python
fp = open(json_file)
json_data = json.load(fp)
fp.close()
print('Image Num: ', len(data))
```

```python
# 读取一张图片
index = random.randint(0, len(json_data)-1)
img_path = os.path.join(img_dir, json_data[index]['img_fn'])
keypoints = json_data[index]['keypoints']
num_kpt = len(keypoints) // 3
img = np.array(cv2.imread(img_path), dtype=np.float32)
print(img_path)
implot = plt.imshow(img[:,:,[2,1,0]] / 256., alpha=0.5)
display_one_img(plt, keypoints, config.limbSeq, config.colors)
```

```python
# 放缩到指定尺寸
trans_resized = Resized(config.num_kpt, config.size)
resized_img, resized_keypoints = trans_resized(img.copy(), keypoints.copy())
implot = plt.imshow(resized_img[:,:,[2,1,0]] / 256., alpha=0.5)
display_one_img(plt, resized_keypoints, config.limbSeq, config.colors)
```

```python
# 随机旋转
trans_rotate = RandomRotate(config.num_kpt, config.max_degree)
rotate_img, rotate_keypoints = trans_rotate(img.copy(), keypoints.copy())
implot = plt.imshow(rotate_img[:,:,[2,1,0]] / 256., alpha=0.5)
display_one_img(plt, rotate_keypoints, config.limbSeq, config.colors)
```

```python
# 随机裁剪
trans_crop = RandomCrop(config.num_kpt, config.ratio_max_x, config.ratio_max_y)
crop_img, crop_keypoints = trans_crop(img.copy(), keypoints.copy())
implot = plt.imshow(crop_img[:,:,[2,1,0]] / 256., alpha=0.5)
display_one_img(plt, crop_keypoints, config.limbSeq, config.colors)
```

```python
# 随机水平翻转
trans_flip = RandomHorizontalFlip(config.num_kpt, config.prob)
flip_img, flip_keypoints = trans_flip(img.copy(), keypoints.copy())
implot = plt.imshow(flip_img[:,:,[2,1,0]] / 256., alpha=0.5)
display_one_img(plt, flip_keypoints, config.limbSeq, config.colors)
```

```python
# 随机颜色干扰
trans_noise = RandomAddColorNoise(config.num_kpt, config.min_gauss, config.max_gauss, config.percentage)
noise_img, noise_keypoints = trans_noise(img.copy(), keypoints.copy())
implot = plt.imshow(noise_img[:,:,[2,1,0]] / 256., alpha=0.5)
display_one_img(plt, noise_keypoints, config.limbSeq, config.colors)
```

```python
# 随机亮度对比度调整
trans_bricon = RandomBrightnessContrast(config.num_kpt, config.min_alpha, config.max_alpha)
bricon_img, bricon_keypoints = trans_bricon(img.copy(), keypoints.copy())
implot = plt.imshow(bricon_img[:,:,[2,1,0]] / 256., alpha=0.5)
display_one_img(plt, bricon_keypoints, config.limbSeq, config.colors)
```

```python
# 多个数据增强一起使用
trans = Compose([
    RandomAddColorNoise(config.num_kpt, config.min_gauss, cfg.max_gauss, cfg.percentage),
    RandomBrightnessContrast(cfg.num_kpt, cfg.min_alpha, cfg.max_alpha),
    RandomCrop(cfg.num_kpt, cfg.ratio_max_x, cfg.ratio_max_y),
    RandomRotate(cfg.num_kpt, cfg.max_degree),
    RandomHorizontalFlip(cfg.num_kpt, cfg.prob),
    Resized(cfg.num_kpt, cfg.input_size),
])
trans_img, trans_keypoints = trans(img.copy(), keypoints.copy())
implot = plt.imshow(trans_img[:,:,[2,1,0]] / 256., alpha=0.5)
display_one_img(plt, trans_keypoints, cfg.limbSeq, cfg.colors)
```
